# Srovnání detectorů obličeje

### Import modulu

```python
import time
from utils.utk_loader import load_utkface_images
from evaluation.evaluation import evaluate
from utils.sum_results import summarize_results
from detectors.mtcnn_detector import MTCNNDetector
from detectors.haar_detector import HaarCascadeDetector
from detectors.scrfd_detector import SCRFDDetector
from detectors.blazeface_detector import BlazeFaceDetector
from detectors.face_recognition_detector import FaceRecognitionDetector
```

```python
dataset = load_utkface_images("./images/UTKFace", max_images=23708)
```

### OpenCV Haar detector

```python
start_time = time.time()
print("Starting Haar Cascade evaluation...")
detector_haar = HaarCascadeDetector()
stats = evaluate(detector_haar, dataset)
# full result dataframe in results directory
summarize_results(stats, 'HAAR')
end_time = time.time()
print(f"Haar Cascade evaluation took {end_time - start_time:.2f} seconds.")
```

### MTCNN detector

```python
start_time = time.time()
print("Starting MTCNN evaluation...")
detector_mtcnn = MTCNNDetector()
stats = evaluate(detector_mtcnn, dataset)
# full result dataframe in results directory
summarize_results(stats, 'MTCNN')
end_time = time.time()
print(f"MTCNN evaluation took {end_time - start_time:.2f} seconds.")
```

### BLAZEFACE detector

```python
start_time = time.time()
detector_blazeface = BlazeFaceDetector(model_sel=0)
stats = evaluate(detector_blazeface, dataset)
# full result dataframe in results directory
summarize_results(stats, 'BLAZEFACE')
end_time = time.time()
print(f"Blazeface evaluation took {end_time - start_time:.2f} seconds.")
```

### SCRFD detector 

```python

start_time = time.time()
detector_scrfd = SCRFDDetector()
print("Starting SCRFD evaluation...")
stats = evaluate(detector_scrfd, dataset)
# full result dataframe in results directory
summarize_results(stats, 'SCRFD')
end_time = time.time()
print(f"SCRFD evaluation took {end_time - start_time:.2f} seconds.")
```

# Face recognition detector

```python
start_time = time.time()
detector_face = FaceRecognitionDetector()
print("Starting FaceRecognition evaluation...")
stats = evaluate(detector_face, dataset)
# full result dataframe in results directory
summarize_results(stats, 'FaceRecognition')
end_time = time.time()
print(f"FaceRecognition evaluation took {end_time - start_time:.2f} seconds.")
```
