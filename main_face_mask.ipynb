# Face mask dataset

## Importu modulu

```python
import time
from evaluation.eval_face_mask import evaluate_detector
from utils.sum_results_mask import summarize_mask_results
from utils.face_mask_loader import load_face_mask_data
from detectors.mtcnn_detector import MTCNNDetector
from detectors.haar_detector import HaarCascadeDetector
from detectors.scrfd_detector import SCRFDDetector
from detectors.blazeface_detector import BlazeFaceDetector
from detectors.face_recognition_detector import FaceRecognitionDetector
import kagglehub
```

```python
path = kagglehub.dataset_download("andrewmvd/face-mask-detection")
dataset = load_face_mask_data(path, max_images=853)
```

### Haar detector

```python
start_time = time.time()
print("Starting Haar Cascade evaluation...")
detector_haar = HaarCascadeDetector()
stats = evaluate_detector(detector_haar, dataset)
summarize_mask_results(stats, 'HAAR')
end_time = time.time()
print(f"Haar Cascade evaluation took {end_time - start_time:.2f} seconds.")
```

### MTCNN detector

```python
start_time = time.time()
print("Starting MTCNN evaluation...")
detector_mtcnn = MTCNNDetector()
stats = evaluate_detector(detector_mtcnn, dataset)
summarize_mask_results(stats, 'MTCNN')
end_time = time.time()
print(f"MTCNN evaluation took {end_time - start_time:.2f} seconds.")
```

###  Blazeface detector

```python
start_time = time.time()
print("Starting Blazeface evaluation...")
detector_blaze = BlazeFaceDetector(model_sel=1)
stats = evaluate_detector(detector_blaze, dataset)
summarize_mask_results(stats, 'BLAZEFACE')
end_time = time.time()
print(f"Blazeface evaluation took {end_time - start_time:.2f} seconds.")
```

### SCRFD detector

```python
start_time = time.time()
print("Starting SCRFD evaluation...")
detector_scrfd = SCRFDDetector()
stats = evaluate_detector(detector_scrfd, dataset)
summarize_mask_results(stats, 'SCRFD')
end_time = time.time()
print(f"SCRFD evaluation took {end_time - start_time:.2f} seconds.")
```

### Face recognition detector

```python
start_time = time.time()
print("Starting Face recognition evaluation...")
detector_face = FaceRecognitionDetector()
stats = evaluate_detector(detector_face, dataset)
summarize_mask_results(stats, 'FACE')
end_time = time.time()
print(f"Face recognition evaluation took {end_time - start_time:.2f} seconds.")
```
